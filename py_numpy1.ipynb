```python
# 2D array

import numpy as np

l2 = np.array([[2, 3, 4, 5], [4, 5, 2, 6], [6, 7, 4, 3]])
print(l2)
```

```python
# Shape & Info Functions


# np.shape() → Shape of array
# np.size() → Total elements
# np.ndim() → Dimensions
# np.dtype() → Data type

print(l2.shape)
print(l2.ndim)
print(l2.size)
print(l2.dtype)
print(l2.itemsize)
```

```python
# Statistical Functions


# np.mean() → Average
# np.median() → Middle value
# np.std() → Standard deviation
# np.var() → Variance
# np.min() → Minimum
# np.max() → Maximum
# np.sum() → Total
# np.argmin() → Index of min
# np.argmax() → Index of max

print("sum : ", np.sum(l2))
print("std : ", np.std(l2))
print("var : ", np.var(l2))
print("min : ", np.min(l2))
print("max : ", np.max(l2))
print("mean : ", np.mean(l2))
print("median : ", np.median(l2))
print("argmin : ", np.argmin(l2))
print("argmax : ", np.argmax(l2))
```
