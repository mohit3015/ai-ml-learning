```python
# Conditional Programming

age = input("Enter your age : ")

if int(age) >= 18:
  print("You are eligible for voting")

else:
  print("You are not eligible for voting")
```

```python
marks = int(input("Enter your marks : "))

if (marks >= 35):
  print("You pass")

elif (marks >= 60):
  print("You pass with first class")

elif (marks >= 80):
  print("You pass with distinction")

else:
  print("You fail!")
```

```python
numbers = [1, 2, 3, 4, 5, 7]

for num in numbers:
  if num % 2 == 0:
    print(num,"is Even")
  else:
    print(num, "is Odd")
```

```python
attempt = 0
max_attempts = 3

while attempt < max_attempts:
  remaining = max_attempts - attempt - 1

  user_name = input("Enter your name : ")
  password = input("Enter your password : ")

  if (user_name == "Mohit" and password == "Mk@123"):
    print("login sucussfully!")
    break

  else:
    attempt += 1
    if remaining > 0:
      print(f"Login failed | You have {remaining} attempt(s) left")

    else:
      print("Try again after 24 hours")
```
