```python
# Treating missing values in a data frame
```

```python
print("Percentage of missing value - ")
print()
print((df.isna().sum()/len(df))*100)
```

```python
# Cabin column has almost 77% values are missing, which mean this column is having majority of values missing
# it make this column a less important column for this data
# so we can remove this data column

# Drop -> to specific row or a column

df1 = df.drop(columns = ['Cabin'])
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
# Embarked column has just 2 values missing - removing these two rows will not much affect the data
# Dropna() - used to drop all the rows with missing values
df2 = df1.dropna(subset = ['Embarked'])
df2.head(3)
```

```python
df2.isna().sum()
```

```python
# Age => it can be an important dta column with some important insights about data
# We have more then 80% values are known
# We can calculate mean or median to fill the blank spots

print(df2['Age'].mean())
print(df2['Age'].median())
```

```python
# fillnal => to fill the missing values existing in a data column

df2['Age'] = df2['Age'].fillna(29)
```
